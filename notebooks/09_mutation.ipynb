## Test Chromosome

```python
import numpy as np
import seaborn as sns

PALETTE = "Set2"
SEED = 37

%matplotlib inline
sns.set(style='white', context='notebook', palette=PALETTE, rc={'figure.figsize':(14,10)})

rng = np.random.default_rng(SEED)

tour = np.array([0, 1, 2, 3, 4, 5, 6, 7])

print("Original tour:", tour)
```

## Helper Function for Feasibility Check

```python
def is_valid_permutation(chromosome, n_cities):
    """Return True if chromosome is a valid permutation of 0..n_cities-1."""
    return (
        len(chromosome) == n_cities
        and len(set(chromosome)) == n_cities
        and set(chromosome) == set(range(n_cities))
    )
```

## Swap Mutation

Swap mutation selects two positions in the chromosome and exchanges their values.  
Since it only swaps existing cities, it does not create duplicates or missing cities.

### Algorithm

1. Copy the original chromosome.
2. Generate a random number.
3. If the random number is less than the mutation rate, select two different positions.
4. Swap the cities at those positions.
5. Return the mutated chromosome.

### Implementation

```python
def swap_mutation(chromosome, rate, rng):
    """
    Perform swap mutation on a chromosome.

    chromosome: 1-D array representing a TSP tour.
    rate      : probability of applying mutation.
    rng       : NumPy random generator.
    """
    mutated = chromosome.copy()

    if rng.random() < rate:
        i, j = rng.choice(len(mutated), size=2, replace=False)
        mutated[i], mutated[j] = mutated[j], mutated[i]

    return mutated
```

### Worked Example

```python
rng_demo = np.random.default_rng(42)

swap_child = swap_mutation(tour, rate=1.0, rng=rng_demo)

print("Before swap mutation:", tour)
print("After swap mutation :", swap_child)
print("Valid permutation   :", is_valid_permutation(swap_child, len(tour)))
```

### Unit Test (1000 random mutations)

```python
rng_test = np.random.default_rng(123)
n = 20

for trial in range(1000):
    parent = rng_test.permutation(n)
    child = swap_mutation(parent, rate=0.5, rng=rng_test)

    assert is_valid_permutation(child, n), f"Invalid swap mutation on trial {trial}: {child}"

print("Swap mutation: 1000/1000 trials produced valid permutations.")
```

## Summary

Swap mutation randomly selects two different positions in a chromosome and swaps the cities at those positions. The mutation rate controls whether the swap is applied during each call.

Because swap mutation only rearranges existing cities, it does not create duplicate or missing cities. Therefore, the output remains a valid permutation for the TSP chromosome.
