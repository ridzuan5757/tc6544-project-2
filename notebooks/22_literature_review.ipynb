## Genetic Algorithms for the TSP

Maps to `Lit:GAforTSP` in `LiteratureReview.tex`.

- Holland (1975) and Goldberg (1989) provide the foundational GA framing: population-based search, selection, crossover, and mutation.
- Reinelt (1991) supports the use of TSPLIB as the benchmark source for instances such as kroA100.
- Larranaga et al. (1999) is the key classic GA-TSP survey for representations and operators.
- Scholz (2019), `literature_reviews/scholz-2019-ga-tsp-historical-review.md`, frames the historical development of GA-TSP work and recurring issues such as local search and premature convergence.
- Liu (2014), `literature_reviews/liu-2014-powerful-genetic-algorithm-tsp.md`, shows why edge structure and local search matter for strong GA-TSP performance.
- Eremeev and Kovalenko (2017), `literature_reviews/eremeev-2017-optimal-recombination-atsp.md`, provides supporting evidence that recombination design and diversity management affect TSP-family GA performance.


## Constraint Handling in Evolutionary Algorithms

Maps to `Lit:ConstraintHandling`.

- Coello Coello (2002) is the central survey for constraint-handling techniques in evolutionary algorithms.
- Michalewicz and Schoenauer (1996) supports the argument that constraints must be reflected in representation, operators, repair, or evaluation.
- Chehouri et al. (2016), `literature_reviews/chehouri-2016-constraint-handling-violation-factor-ga.md`, critiques penalty-function tuning and motivates alternative feasibility-aware selection strategies.
- Padhye, Mittal, and Deb (2015), `literature_reviews/padhye-2015-feasibility-preserving-constraint-handling.md`, discusses feasibility-preserving repair strategies and the risk that repair can bias search or reduce exploration.

Project connection: the TSP constraint is not a minor side condition. A chromosome that repeats a city and omits another city is not a valid tour, and direct fitness evaluation can produce misleadingly low distances.


## Repair Mechanisms in Permutation-based EAs

Maps to `Lit:Repair`.

- Salcedo-Sanz (2009) is the most directly relevant repair survey and should be used with Coello Coello (2002) and Michalewicz and Schoenauer (1996).
- Padhye et al. (2015) supports the critical analysis that repair restores feasibility but may change the search distribution.
- Cicirello (2023), `literature_reviews/cicirello-2023-survey-evolutionary-operators-permutations.md`, explains why permutation operators should be understood by the features they preserve: positions, edges, directed edges, precedences, and cyclic precedences.
- Uray et al. (2023), `literature_reviews/uray-2023-csrx-crossover-tsp.md`, supports the distinction between feasibility and quality: a valid tour can still lose useful edge or symmetry structure.

Project connection: duplicate-removal plus missing-city insertion is appropriate for making offspring feasible, but its effect on diversity and convergence must be measured rather than assumed.


## Comparison Studies of Crossover Operators (PMX/OX/CX)

Supports `T2:Crossover` in `Task2.tex` and the literature-review subsection `Lit:CrossoverStudies`.

- Goldberg and Lingle (1985), Davis (1985), and Oliver et al. (1987) support the classic PMX, OX, and CX operator discussion.
- Abdoun and Abouchabaka (2012), `literature_reviews/abdoun-2012-adaptive-crossover-operators-tsp.md`, compares TSP crossover operators and supports the claim that OX/PMX/CX-style choices materially affect GA performance.
- Abdoun, Abouchabaka, and Tajani (2012), `literature_reviews/abdoun-2012-mutation-operators-tsp.md`, supports including mutation choice and mutation rate in GA design.
- Thanh, Binh, and Lam (2020), `literature_reviews/thanh-2020-combination-crossover-ga-tsp.md`, uses TSPLIB instances including kroA100-sized problems, making it useful for project-scale context.
- Cicirello (2023) gives a modern taxonomy of permutation crossover and mutation operators.
- Uray et al. (2023) adds a modern symmetry-aware crossover perspective.
